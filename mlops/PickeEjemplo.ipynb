```python
import joblib
import pandas as pd

print("Iniciando prueba local del modelo...")

# 1. Cargar los artefactos empaquetados
try:
    artefactos = joblib.load('mejor_modelo_california.pkl')
    modelo = artefactos['modelo']
    escalador = artefactos['escalador']
    print("✅ Modelo y escalador cargados correctamente.\n")
except FileNotFoundError:
    print("❌ Error: No se encontró el archivo 'mejor_modelo_california.pkl'. Asegúrate de que esté en esta carpeta.")
    exit()

# 2. Definir los datos hardcodeados (Una casa de ejemplo en California)
# Es importante usar un DataFrame con los nombres de las columnas para que el escalador
# no arroje advertencias (warnings) por falta de nombres de características.
columnas = [
    "MedInc", "HouseAge", "AveRooms", "AveBedrms",
    "Population", "AveOccup", "Latitude", "Longitude"
]

# Puedes cambiar estos valores para probar diferentes resultados
datos_casa_ejemplo = [[8.32, 41.0, 6.98, 1.02, 322.0, 2.55, 37.88, -122.23]]

df_prueba = pd.DataFrame(datos_casa_ejemplo, columns=columnas)

print("📊 Datos de la casa a evaluar:")
print(df_prueba.to_string(index=False))
print("-" * 40)

# 3. Preprocesar (Escalar) los datos
# Usamos .transform() para aplicar la misma escala matemática que se usó en el entrenamiento
datos_escalados = escalador.transform(df_prueba)

# 4. Generar la predicción
prediccion = modelo.predict(datos_escalados)

# 5. Mostrar el resultado
# El dataset de California Housing expresa el target en cientos de miles de dólares.
# Multiplicamos por 100 para mostrarlo en "miles de dólares" y que sea más legible.
precio_estimado = prediccion[0] * 100

print(f"\n🏠 PRECIO ESTIMADO: ${precio_estimado:,.2f} miles de dólares")
print("¡Prueba exitosa!")
```
